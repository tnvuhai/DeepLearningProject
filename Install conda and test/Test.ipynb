```python
import cv2
import imutils
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
image = cv2.imread("C:\\Users\\mamto\\Desktop\\HuySmile.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#image = imutils.resize(image, 150)
plt.imshow(image)
plt.colorbar()
```

```python
grayscale = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
gray = cv2.cvtColor(grayscale, cv2.COLOR_BGR2RGB)
fig, aes = plt.subplots(1,2, figsize = (15,10))
aes[0].imshow(image)
aes[0].set_title("Huy smile original")

aes[1].imshow(gray)
aes[1].set_title("Huy smile grayscale")
plt.show()
```

```python
from matplotlib.pyplot import figure

faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(100, 100),
    flags = cv2.CASCADE_SCALE_IMAGE
)
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
#cv2.waitKey(0)
```

```python
import cv2
import sys

cascPath = sys.argv[1]
faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
video_capture = cv2.VideoCapture('Video.mp4')

while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(10, 10),
        flags=cv2.CASCADE_SCALE_IMAGE
    )

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Display the resulting frame
    cv2.imshow('Video', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything is done, release the capture
video_capture.release()
cv2.destroyAllWindows()
```
